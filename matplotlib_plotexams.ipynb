```python
import matplotlib.pyplot as plt
import numpy as np

### 기본 플롯 구조
'''
fig: 전체 그래프 창
axes: 실제 그래프가 그려지는 영역
title: 그래프 제목
xlabel/ylabel: 축 레이블
legend: 범례
'''
```

```python
# 데이터 준비
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 기본 라인 플롯
plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.plot(x, y, 'b-', label='sin(x)')  # 파란색 실선
plt.title('Basic Line Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)  # 그리드 추가
plt.legend()
plt.show()
```

```python
# 여러 선 스타일과 마커
plt.figure(figsize=(12, 8))
plt.plot(x, np.sin(x), 'r--', label='sin(x)')  # 빨간색 점선
plt.plot(x, np.cos(x), 'b:', label='cos(x)')   # 파란색 점선
plt.plot(x, -np.sin(x), 'g.-', label='-sin(x)') # 초록색 선+점
plt.title('Multiple Lines')
plt.legend()
plt.grid(True)
plt.show()

'''
선 스타일: - (실선), -- (점선), : (점선), -. (대시점선)
색상: r (빨강), g (초록), b (파랑), k (검정), w (흰색)
마커: . (점), o (원), s (사각형), ^ (삼각형)
'''
```

```python
# 2x2 서브플롯
plt.figure(figsize=(12, 8))

# 첫 번째 서브플롯
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title('Sine')

# 두 번째 서브플롯
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.title('Cosine')

# 세 번째 서브플롯
plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x))
plt.title('Tangent')

# 네 번째 서브플롯
plt.subplot(2, 2, 4)
plt.plot(x, np.exp(x))
plt.title('Exponential')

plt.tight_layout()  # 서브플롯 간격 자동 조정
plt.show()
```

```python
# 랜덤 데이터 생성
np.random.seed(42)
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 1.5, 100)

plt.figure(figsize=(10, 6))
plt.scatter(data1, data2, c='blue', alpha=0.5, label='Data Points')
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 데이터 준비
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True, axis='y')
plt.show()

# 수평 막대 그래프
plt.figure(figsize=(10, 6))
plt.barh(categories, values, color='lightgreen')
plt.title('Horizontal Bar Plot')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.grid(True, axis='x')
plt.show()
```

```python
# 정규분포 데이터 생성
data = np.random.normal(0, 1, 1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 데이터 준비
sizes = [30, 20, 25, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']
colors = ['red', 'green', 'blue', 'yellow', 'orange']

plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.axis('equal')  # 원형 유지
plt.show()
```

```python
# 여러 그룹의 데이터 생성
data = [np.random.normal(0, std, 100) for std in [1, 2, 3]]

plt.figure(figsize=(10, 6))
plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('Box Plot')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
# 스타일 설정
plt.style.use('seaborn')  # 'ggplot', 'dark_background' 등

# 텍스트 속성 설정
plt.figure(figsize=(10, 6))
plt.plot(x, np.sin(x))
plt.title('Styled Plot', fontsize=16, fontweight='bold')
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

# 축 범위 설정
plt.xlim(-2, 12)
plt.ylim(-1.5, 1.5)

# 주석 추가
plt.annotate('Maximum', xy=(1.5, 1), xytext=(2, 1.2),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```
