```python
'''
그림판으로 만들었던 컬러 이미지를 파이썬 코드를 이용하여 생성한다.
'''
colorImg = np.zeros((10,10,3)).astype('uint8')
colorImg[:5, :5, :] = 255 # white
colorImg[:5, 5:, 0] = 255 # blue
colorImg[5:, :5, 1] = 255 # green
colorImg[5:, 5:, 2] = 255 # red

print('sum of color image:', np.sum(colorImg))

upsampled = cv2.resize(colorImg, (500, 500), interpolation=cv2.INTER_LINEAR)
print('sum of upsampled color image:', np.sum(upsampled))
cv2.imwrite('sample_data/upsampledColorImg.jpg',upsampled)
```

```python
'''
세로 0~249, 가로 0~249 까지 흰색
세로 0~249, 가로 250~499 까지 파란색
세로 250~499, 가로 0~250 까지 초록색
세로 250~499, 가로 250~499 까지 빨간색
을 갖는 500x500x3의 컬러 이미지를 반복문을 이용하여 생성한다.
'''
resImg500 = np.zeros((500,500,3)).astype('uint8')
for i in range(500):
  for j in range(500):
    for k in range(3):
      if i < 250:
        if j < 250:
          resImg500[i,j,k] = 255
        else:
          if k == 0: resImg500[i,j,k] = 255
      else:
        if j< 250:
          if k == 1:
            resImg500[i,j,k] = 255
        else:
          if k == 2:
            resImg500[i,j,k] = 255

cv2_imshow(resImg500)
print('total WBGR pixels of resolution 500x500:',sum(resImg500.reshape(500*500*3)))
```

```python
grayImg = np.zeros((10,10)).astype('uint8')

for i in range(10):
  for j in range(10):
    grayImg[i,j] += round(colorImg[i,j,0]*0.114)
    grayImg[i,j] += round(colorImg[i,j,1]*0.587)
    grayImg[i,j] += round(colorImg[i,j,2]*0.299)

cv2.imwrite('grayImg.jpg',grayImg)
print(grayImg)
```

```python
'''
dissolve example
'''
soccer_img = cv2.imread('sample_data/soccer.jpg') # soccer image read
mistry_img = cv2.imread('sample_data/mistyroad.jpg') # mistryroad image read
print('Soccer image')
cv2_imshow(cv2.resize(soccer_img, (0, 0), fx=0.25, fy=0.25, interpolation=cv2.INTER_NEAREST))
print()
print('Mistryroad image')
cv2_imshow(cv2.resize(mistry_img, (0, 0), fx=0.25, fy=0.25, interpolation=cv2.INTER_NEAREST))

resized_soccer_img = cv2.resize(soccer_img, (200,200))
resized_mistry_img = cv2.resize(mistry_img, (200,200))
for sigma in range(0, 11, 2):
  sigma /= 10
  dissolve_img = resized_soccer_img * sigma + resized_mistry_img * (1-sigma)
  dissolve_stack = dissolve_img if sigma == 0 else np.hstack((dissolve_stack, dissolve_img))
print('stacked dissolve images')
cv2_imshow(dissolve_stack)
```

```python
'''
Contrast edit example
'''
contrast_test_img = resized_mistry_img.astype(np.int32)
for cont in range(-120, 160, 40):
  edit_img = contrast_test_img + cont
  edit_img = np.clip(edit_img, 0, 255)
  contrast_strack = edit_img if cont == -120 else np.hstack((contrast_strack, edit_img))
print('Stacked contrast changed images')
cv2_imshow(contrast_strack)
```

```python
def gamma(frame, gamma=1.0):
  new_frame = frame/255.0
  return np.uint8(255*(new_frame**gamma))

gamma_stack = np.hstack((gamma(resized_soccer_img, 0.5),
                         gamma(resized_soccer_img, 0.75),
                         gamma(resized_soccer_img, 1.0),
                         gamma(resized_soccer_img, 2.0),
                         gamma(resized_soccer_img, 3.0)))

print('Straced gamma changed images')
cv2_imshow(gamma_stack)
```

```python
def cal_convolution(frame, filter3x3):
  convolution_img = np.zeros((height, width))
  for j in range(1, height-1):
    for i in range(1, width-1):
      convolution_img[j, i] = filter3x3[0, 0]*frame[j-1,i-1] + \
                              filter3x3[0, 1]*frame[j-1,i] + \
                              filter3x3[0, 2]*frame[j-1,i+1] + \
                              filter3x3[1, 0]*frame[j,i-1] + \
                              filter3x3[1, 1]*frame[j,i] + \
                              filter3x3[1, 2]*frame[j,i+1] + \
                              filter3x3[2, 0]*frame[j+1,i-1] + \
                              filter3x3[2, 1]*frame[j+1,i] + \
                              filter3x3[2, 1]*frame[j+1,i+1]
  return convolution_img

gray_soccer_img = cv2.cvtColor(resized_soccer_img, cv2.COLOR_BGR2GRAY)
height, width = gray_soccer_img.shape
# gaussian filter | sigma=1.0
smoothing_filter = np.array([[0.0751, 0.1238, 0.0751],
                             [0.1238, 0.2042, 0.1238],
                             [0.0751, 0.1238, 0.0751]])

box_filter = np.array([[1/9, 1/9, 1/9],
                       [1/9, 1/9, 1/9],
                       [1/9, 1/9, 1/9]])

hedge_filter = np.array([[-1, 0, 1],
                         [-1, 0, 1],
                         [-1, 0, 1]])

vedge_filter = np.array([[-1, -1, -1],
                         [0, 0, 0],
                         [1, 1, 1]])

print('Stacked filtered images')
convolution_stack = np.hstack((gray_soccer_img,
                               cal_convolution(gray_soccer_img, smoothing_filter),
                               cal_convolution(gray_soccer_img, box_filter),
                               cal_convolution(gray_soccer_img, hedge_filter),
                               cal_convolution(gray_soccer_img, vedge_filter)))
cv2_imshow(convolution_stack)
```
