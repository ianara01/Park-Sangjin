Colab을 통해 인기 있는 Python 라이브러리를 최대한 활용하여 데이터를 분석하고 시각화할 수 있습니다. 아래 코드 셀에서는 <strong>Numpy</strong>를 사용하여 임의의 데이터를 생성하고 <strong>매트플롯립</strong>으로 이를 시각화합니다. 셀을 클릭하면 코드 수정을 바로 시작할 수 있습니다.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
from pymongo import MongoClient

# 1. MongoDB 연결 생성
def connect_to_mongo():
    try:
        client = MongoClient("mongodb://admin:1234@hanslab.org:57017")  # MongoDB 연결 (기본 로컬 주소와 포트)
        print("MongoDB 연결 성공")
        return client
    except Exception as e:
        print(f"MongoDB 연결 실패: {e}")
        return None

# 2. Database 생성
def create_database(client, db_name):
    db = client[db_name]
    print(f"Database '{db_name}' 생성 또는 연결 완료")
    return db

# 3. Collection 생성 및 데이터 입력
def create_and_insert_collection(db, collection_name, data):
    collection = db[collection_name]
    if isinstance(data, list):
        result = collection.insert_many(data)  # 여러 문서 삽입
        print(f"{len(result.inserted_ids)}개의 문서가 삽입되었습니다.")
    else:
        result = collection.insert_one(data)  # 단일 문서 삽입
        print(f"문서 삽입 완료: ID={result.inserted_id}")
    return collection

# 4. 데이터 조회 함수
def fetch_all_documents(collection):
    documents = list(collection.find())  # 모든 문서 조회
    return documents

# 5. 실행 예시
if __name__ == "__main__":
    # MongoDB 연결
    client = connect_to_mongo()
    if not client:
        exit()  # 연결 실패 시 종료

    # Database 생성
    db_name = "likelion"
    db = create_database(client, db_name)

    # Collection 생성 및 데이터 삽입
    collection_name = "students"
    sample_data = [
        {"name": "Alice", "age": 25, "major": "Computer Science", "email": "ianara01@naver.com"},
        {"name": "Bob", "age": 22, "major": "Mechanical Engineering"},
        {"name": "Charlie", "age": 23, "major": "Mathematics"},
    ]
    collection = create_and_insert_collection(db, collection_name, sample_data)

    # 데이터 조회
    documents = fetch_all_documents(collection)
    print("조회된 문서:")
    for doc in documents:
        print(doc)

    # MongoDB 연결 종료
    client.close()
```
